1: Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)

    if parked > capacity:
        print("Alert: Parking Exceeded Capacity!")
    else:
        print("Parking Status: Available")


# Sample Input
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 mins per km

    # Traffic delay
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10

    # Weather delay
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
seat_occupancy(200, [1]*150)
```

4: Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(avg_cpu), "%")

    if avg_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load([45, 60, 70, 85, 90])
```

5: Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:   # Threshold = 8 hours
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
