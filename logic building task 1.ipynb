# 1. User Login Check (Very Simple Version)


```python
# User Login Check Program

# Correct username and password
correct_username = "admin"
correct_password = "1234"

# Ask user to enter username and password
username = input("Enter username: ")
password = input("Enter password: ")

# Check if both are correct
if username == correct_username:
    if password == correct_password:
        print("Login Successful")
    else:
        print("Invalid Credentials")
else:
    print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer (Simple Loop Version)


```python
# Pass / Fail Analyzer Program

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Check each mark
for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

print("Total students:", len(marks))
print("Number of Pass students:", pass_count)
print("Number of Fail students:", fail_count)
```

# 3. Simple Data Cleaner (Basic Method)


```python
# Simple Data Cleaner Program

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    name = name.strip()      # Remove spaces
    name = name.lower()      # Convert to lowercase
    cleaned_names.append(name)

print("Cleaned Names:", cleaned_names)
```

# 4. Message Length Analyzer (Easy Version)


```python
# Message Length Analyzer Program

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)
    
    if length > 10:
        print("This message is longer than 10 characters")
    
    print()   # blank line
```

# 5. Error Message Detector (Basic Counting)


```python
# Error Message Detector Program

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count = 0

for log in logs:
    if log == "ERROR":
        count = count + 1

print("Total ERROR messages:", count)
```
